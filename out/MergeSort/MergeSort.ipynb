## 1. Introduction

Merge Sort is a divide and conquer based sorting algorithm.
It works by dividing the array into smaller subarrays, sorting them recursively, and then merging the sorted subarrays to produce the final sorted array.

Merge Sort is known for its stable sorting behavior and guaranteed time complexity.

## 3. Algorithm (Step-by-Step)

#### Step 1: Divide

If the array has more than one element:

Divide the array into two halves

#### Step 2: Conquer

Recursively apply Merge Sort on:

Left subarray

Right subarray

#### Step 3: Merge

Compare elements of both sorted halves

Merge them into a single sorted array

## 4. Pseudocode

```python
MERGE_SORT(arr):
    if length of arr == 1
        return

    divide arr into left and right

    MERGE_SORT(left)
    MERGE_SORT(right)

    MERGE(left, right, arr)
```

## 5. Java Implementation (Recursive)

```python
public class MergeSort {

    public static void main(String[] args) {
        int[] arr = {5, 2, 8, 4, 1, 6, 7, 3, -5};

        mergeSort(arr);

        for (int ele : arr) {
            System.out.print(ele + " ");
        }
    }

    private static void mergeSort(int[] arr) {
        int n = arr.length;
        if (n == 1) return;

        int[] a = new int[n / 2];
        int[] b = new int[n - n / 2];

        int idx = 0;
        for (int i = 0; i < a.length; i++) {
            a[i] = arr[idx++];
        }
        for (int i = 0; i < b.length; i++) {
            b[i] = arr[idx++];
        }

        mergeSort(a);
        mergeSort(b);

        merge(a, b, arr);
    }

    public static void merge(int[] a, int[] b, int[] c) {
        int i = 0, j = 0, k = 0;

        while (i < a.length && j < b.length) {
            if (a[i] <= b[j]) {
                c[k++] = a[i++];
            } else {
                c[k++] = b[j++];
            }
        }

        while (i < a.length) {
            c[k++] = a[i++];
        }

        while (j < b.length) {
            c[k++] = b[j++];
        }
    }
}
```
