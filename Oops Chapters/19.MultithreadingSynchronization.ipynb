## ðŸŸ¥ What is Synchronization? (Theory)

In Java multithreading, synchronization is a mechanism that controls the access of multiple threads to shared resources.

When two or more threads try to access the same shared resource at the same time, there is a chance of data inconsistency, race conditions, and unexpected outputs.

To prevent this problem, Java provides synchronization, which ensures that only one thread can access a critical section at a time.

### ðŸŸ¦ 1. Synchronized Method

In synchronized method, the entire method is locked, and only one thread can execute it at a time.

âœ”ï¸ Example: Bank Withdrawal
Theory

A synchronized method is used when the whole method contains critical code. The lock is applied on the object.

Code

```java
class Bank {
    private int balance = 10000;

    synchronized void withdraw(int amount) {
        System.out.println(Thread.currentThread().getName() + " is trying to withdraw " + amount);

        if (balance >= amount) {
            System.out.println("Withdrawal successful!");
            balance -= amount;
        } else {
            System.out.println("Insufficient balance!");
        }

        System.out.println("Remaining Balance: " + balance);
    }
}

public class SyncMethodDemo {
    public static void main(String[] args) {
        Bank obj = new Bank();

        Thread t1 = new Thread(() -> obj.withdraw(8000), "Person A");
        Thread t2 = new Thread(() -> obj.withdraw(7000), "Person B");

        t1.start();
        t2.start();
    }
}
```

### ðŸŸ© 2. Synchronized Block
Theory

Instead of locking the whole method, you can lock only a specific block of code.
This gives better performance.

Code

```java
class Table {
    void printTable(int n) {
        synchronized(this) {
            for (int i = 1; i <= 5; i++) {
                System.out.println(n * i);
            }
        }
    }
}

public class SyncBlockDemo {
    public static void main(String[] args) {
        Table t = new Table();

        Thread t1 = new Thread(() -> t.printTable(5));
        Thread t2 = new Thread(() -> t.printTable(10));

        t1.start();
        t2.start();
    }
}
```

### ðŸŸ¦ 3. Static Synchronization
Theory

Static synchronized methods lock on the class object, not the instance.
Used when multiple objects share the same data.

Code

```java
class Display {
    synchronized static void showNumbers(int n) {
        for (int i = 1; i <= 5; i++) {
            System.out.println(n * i);
        }
    }
}

public class StaticSyncDemo {
    public static void main(String[] args) {

        Thread t1 = new Thread(() -> Display.showNumbers(2));
        Thread t2 = new Thread(() -> Display.showNumbers(3));

        t1.start();
        t2.start();
    }
}
```

### ðŸŸ§ 4. Inter-thread Communication (wait, notify)
Theory

Used when threads must communicate with each other. Example:

One thread produces data

Another thread consumes data

Code

```java
class Data {
    int value;
    boolean flag = false;

    synchronized void produce(int v) {
        try {
            if (flag) wait();  
            value = v;
            System.out.println("Produced: " + value);
            flag = true;
            notify();
        } catch (Exception e) {}
    }

    synchronized void consume() {
        try {
            if (!flag) wait();
            System.out.println("Consumed: " + value);
            flag = false;
            notify();
        } catch (Exception e) {}
    }
}

public class IPCDemo {
    public static void main(String[] args) {
        Data d = new Data();

        Thread producer = new Thread(() -> {
            for (int i = 1; i <= 5; i++)
                d.produce(i);
        });

        Thread consumer = new Thread(() -> {
            for (int i = 1; i <= 5; i++)
                d.consume();
        });

        producer.start();
        consumer.start();
    }
}
```

#### ðŸŸ¨ What is a Lock?

Synchronization uses a monitor lock.
Only one thread can hold the lock and execute synchronized code.

#### ðŸŸª Thread Safety

If a class behaves correctly when accessed by multiple threads simultaneously, it is thread-safe.

Synchronization helps achieve thread safety.
